## 1. Calculation of new fields

The creation of new fields from fields already existing in the data set is a very simple task.Let's illustrate this procedure using what is perhaps the most used package of Python language, *pandas *.Pandas is a software library written for the Python programming language for data handling and analysis.In particular, it offers data structures and operations to manipulate numerical tables and time series.We will use the "Invoice.csv" data set to illustrate the creation of new fields.

```python
# Importing the dataset
invoices = pd.read_csv('/workspaces/Auditoria/An√°lises_e_Testes_Comuns_da_auditoria_utilizando_Python/Data/Invoices.csv', sep =";")
invoices.head()
```

```python
invoices.dtypes # Shows the type of each column of the data set
```

Here we see that the 'unitprice' and 'amount' columns are of the pandas object type, which means that he is understanding the fields as 'string', text, so it will not be possible to do mathematical operations that are part of this demonstration, for that,We will convert these two columns.

```python
invoices['UnitPrice'] = invoices['UnitPrice'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('float')
invoices['Amount'] = invoices['Amount'].apply(lambda x: x.replace('$', '').replace(',', '')).astype('float')
invoices.dtypes
```

As we see now, the two fields have been converted.

Let's create a new field in the Invoice Database called Vlraudit which will be the product of the Quantity and UnitPrice fields.

```python
invoices['VlrAudit'] = invoices['Quantity'] * invoices['UnitPrice']
invoices.head()
```

Now let's create another field we'll call DIF and that will be the difference between the Amount and Vlraudit fields.

```python
invoices['Dif'] = invoices['Amount'] - invoices['VlrAudit']
invoices.head()
```

Now let's create a new field called Vlracum containing the accumulated value of the Amount field.

```python
invoices['VlrAcum'] = invoices['Amount'].cumsum()
invoices.head()
```

Se quisermos deletar as colunas que foram acrescentadas e voltar ao conjunto de dados original, podemos fazer o seguinte:

```python
invoices = invoices.drop(['VlrAudit','Dif','VlrAcum'],axis = 1) # O par√¢metro axis = 1 indica que queremos que a coluna toda 
#seja eliminada.
```

```python
invoices.head()
```

A tabula√ß√£o cruzada consiste normalmente em se determinar a distribui√ß√£o de frequ√™ncias de uma, duas ou mais vari√°veis categ√≥ricas. O Python disp√µe das fun√ß√µes crosstab() e groupby() que nos permitem implementar esta t√©cnica.

## 8. *Gaps* 

Diversos documentos s√£o emitidos seguindo uma numera√ß√£o sequencial; o que permite um maior controle sobre os mesmos. Exemplos disso s√£o a numera√ß√£o sequencial das notas fiscais emitidas por uma empresa, dos cheques emitidos, das notas de empenhos, dos contratos celebrados, etc.

Um recurso muito √∫til e usualmente dispon√≠vel nos softwares de auditoria de uso geral √© a possibilidade de idenficar a exist√™ncia de itens faltantes em um conjunto de dados, ou seja, verificar se existe algum gap (lacuna) na numera√ß√£o sequencial.

Vamos ilustrar como fazer isso usando o conjunto de dados Invoices.csv.

A abordagem utilizada ser√° descrita a seguir:

```python
# Importar arquivos
invoices = pd.read_csv('Invoices.csv', sep =";")
invoices.head()
```

O conjunto de dados ser√° ordenado com base na numera√ß√£o sequencial das faturas e cada elemento ser√° subtra√≠do do seu antecessor. Se n√£o houver falha na numera√ß√£o sequencial, espera-se obter como resultado um vetor contendo apenas 1. Onde o resultado for diferente de 1 haver√° uma falha. Vejamos como fazer isso na pr√°tica.

1. - Ordena√ß√£o do conjunto de dados com base no n√∫mero das faturas

```python
faturas = invoices.sort_values('InvoiceNo')
faturas.head()
```

2. - C√°lculo das diferen√ßas

```python
faturas['differ'] = faturas.InvoiceNo.diff()
faturas.head()
```

3. - Identifica√ß√£o dos registros onde h√° gaps

```python
diferencas = faturas[faturas['differ']!= 1.0].iloc[1:] #iloc aqui √© para eliminarmos a primeira linha, pois o resultado dela √© NaN, 
# j√° que n√£o √© poss√≠vel ubtrairmos a primeira linha pela linha antecessora.
diferencas
```

```python
faturas.iloc[[10,11,12,22,23],:]
```

Como podemos ver, as faturas n¬∫ 20010 est√£o duplicadas, falta a fatura n¬∫ 20011 e a 20023.

## 9. Estratifica√ß√£o

A estratifica√ß√£o consiste em classificar os registros de um arquivo de dados em estratos mutuamente exclusivos. Em auditoria, esta classifica√ß√£o √© feita usualmente com base em valores monet√°rios. O procedimento consiste em se estabelecer faixas de valores e indicar, para cada registro, a que faixa de valor ele pertence.

Para ilustrar a execu√ß√£o deste procedimento, ser√° utilizado o conjunto de dados Invoices.csv.

O primeiro passo para se definir os estratos consiste em se estabelecer as faixas de valores para a vari√°vel a ser utilizada na estratifica√ß√£o nas quais se deseja classificar os registros. Para isso, faz-se necess√°rio conhecer os valores m√°ximo e m√≠nimo do conjunto de dados, o que pode ser feito facilmente com a fun√ß√£o min () e max().

```python
# Importar arquivos
invoices = pd.read_csv('Invoices.csv', sep =";")
invoices.head()
```

```python
# Transformando o campo "Amount" e "UnitPrice" para num√©rico.
invoices['UnitPrice'] = invoices['UnitPrice'].apply(lambda x: x.replace('$', '').replace(',', '.')).astype('float')
invoices['Amount'] = invoices['Amount'].apply(lambda x: x.replace('$', '').replace(',', '.')).astype('float')
```

```python
print('O menor valor Total de fatura √©: ', invoices.Amount.min())
```

```python
print('O maior valor Total de fatura √©: ', invoices.Amount.max())
```

Considerando que os valores das faturas variam entre 5,00 e  13.438,50, desejamos estabelecer 3 estratos (baixos valores, valores medianos e altos valores) poder√≠amos ter os seguintes estratos:

**Estrato 1 -**  5,00 a 1.000,00

**Estrato 2 -**  1.000,01 a 10.000,00

**Estrato 3 -**  10.000,01 a 13.438,50

A defini√ß√£o da quantidade de estratos (3 neste exemplo) e da amplitude de cada faixa √© uma escolha subjetiva do auditor. Deve-se observar que as faixas de valores podem ou n√£o ter a mesma amplitude.

Para realizar a classifica√ß√£o dos registros da base de dados em cada um dos estratos definidos utiliza-sa a fun√ß√£o cut(), como exemplificado a seguir:

```python
invoices['Estratos'] = pd.cut(invoices.Amount,[0.00, 1000.00, 10000.00, 13500.00],
                              labels = ['Baixo','M√©dio','Alto'])

invoices.head()
```

## 11. Combinar Arquivos ( merging )

Em determinadas situa√ß√µes ocorre de os dados necess√°rios √† execu√ß√£o de uma tarefa estarem em arquivos diferentes e √© necess√°rio reun√≠-los em um √∫nico conjunto de dados.

Para exemplificar, considere o caso em que o auditor necessite confirmar com terceiros os valores que comp√µem o saldo da conta Clientes, por exemplo.

Os dados cadastrais, como endere√ßo, c√≥digo de identifica√ß√£o do cliente, CEP, etc, constam de um arquivo e os valores devidos pelos clientes, em outro. Ocorre que o auditor necessita do endere√ßo e dos valores devidos para elaborar uma carta de circulariza√ß√£o. Para isso, √© indispens√°vel que em ambos os arquivos exista um campo que identifique o cliente de forma √∫nica(nesse caso utilizaremos a coluna "account"). Para a execu√ß√£o deste procedimento o python disp√µe da fun√ß√£o pandas.merge(), cuja utiliza√ß√£o ser√° mostrada mais adiante.

Estes dois conjuntos de dados s√£o arquivos texto de formato fixo e, dessa forma, sua importa√ß√£o depende de conhecermos o layout dos arquivos, o que consta do documento Descricao Arquivos Dados_v3.doc que acompanha os conjuntos de dados utilizados.

Com base nas informa√ß√µes contidas naquele documento, a importa√ß√£o dos dados pode ser feita da seguinte forma:

```python
# importar o arquivo de contas a receber
contas_a_receber = pd.read_fwf('Arfile.ASC', sep ="\t", widths=[11, 4, 4, 15, 8], 
                               header = None, 
                               names = ['account', 'division', 'store', 'balance', 'duedate'])
```

```python
contas_a_receber.head()
```

```python
# Carregar o arquivo de endere√ßo dos clientes
endereco = pd.read_fwf('Address.ASC', sep= '\t', widths= [11, 33, 33, 30, 25, 5],
                       header = None, 
                       names = ['account', 'name1', 'name2', 'street', 'cityst', 'zip'])
```

```python
endereco.head()
```

```python
novo_conjunto = pd.merge(contas_a_receber, endereco[['account','name1', 'name2', 'street', 'cityst','zip']],
                         on= 'account', how='inner')
```

```python
novo_conjunto.head()
```

No procedimento realizado acima, tamb√©m conhecido como inner join, o conjunto de dados resultante (no nosso caso novo_conjunto) conter√° apenas as informa√ß√µes relativas √†s contas que existirem em ambos os cojuntos de dados. As contas que estiverem no conjunto de dados contas_receber e n√£o estiverem em endereco n√£o constar√£o da base de dados resultante. O mesmo ocorrendo ao contr√°rio, ou seja, as contas que estiverem no conjunto de dados endereco mas n√£o estiverem em contas_receber tamb√©m n√£o ser√£o apresentados. O inner join mostrar√° a interse√ß√£o das duas bases.
